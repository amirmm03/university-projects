```python
#df['genres'] = df['genres'].str.strip("er").split(',')
df.genres = df['genres'].str.strip("[\']").str.split('\', \'').tolist()
df.genres
```

<div dir = "rtl" style="direction:rtl;line-height:250%; textAlign: right">
<font face="Estedad" size=4 color = 2d2d2e style="textAlign: right">
در ادامه، لیست تمام ژانر‌های ممکن را با گرفتن اجتماع روی سطر‌ها بدست آورید و در یک متغیر با نام مناسب ذخیره کنید. سپس تعداد این ژانر‌ها را نیز چاپ کنید.

</font>    

</div>

```python
allGenres ={};
for x in df.genres:
    allGenres = list(set().union(x,allGenres))
len(allGenres)
```

<div dir = "rtl" style="direction:rtl;line-height:250%; textAlign: right">
<font face="Estedad" size=4 color = 2d2d2e style="textAlign: right">
برای بررسی بهتر میزان تعدد ژانر‌های مختلف، دیتافریمی بسازید که تعداد تکرر هر ژانر را نشان دهد.  سپس  ۱۰ ژانر با بیشترین تکرار را چاپ کنید.
<span style="background-color: #eeeeee; font-size: 1.2rem; padding: 4px 8px; border-radius: 20px;">۲ امتیاز</span>
</font>    
</div>

```python
my_dict={'genre': [],
        'repeats': []}
for genre in allGenres:
    count=0
    for arr in df.genres:
        count+= arr.count(genre)
    my_dict['genre'].append(genre)
    my_dict['repeats'].append(count)

genresdataframe =  pd.DataFrame(my_dict)
genresdataframe.nlargest(n=10, columns=['repeats'])
```

```python
newdf = pd.DataFrame( columns=['author', 'numBooks', 'sumPages','numPublishers','firstBook','mostGenres','avgPopularity','sumRatings'])
newdf['author'] = df['author'].unique()
newdf['numBooks'] = df['author'].value_counts(sort=False).values
newdf['sumPages'] = [df.loc[df['author'] == i]['pages'].sum() for i in newdf.author]
newdf['numPublishers'] = [df.loc[df['author'] == i].publisher.nunique() for i in newdf.author]
newdf['firstBook'] = [df.loc[df['author'] == i]['publish year'].min() for i in newdf.author]

from heapq import nlargest as tmpnlarg
mostlist = []
for i in newdf.author:
    
    mlist = list( [df.loc[df['author'] == i]['genres'].sum() ])
    zzz = dict(zip(*np.unique(mlist, return_counts=True)))
    mostlist.append(tmpnlarg(3, zzz, key = zzz.get))
newdf['mostGenres'] = mostlist

newdf['avgPopularity'] = [df.loc[df['author'] == i]['popularity'].sum() for i in newdf.author]
newdf['avgPopularity'] /= newdf['numBooks']

newdf['sumRatings'] = [df.loc[df['author'] == i]['numRatings'].sum() for i in newdf.author]

newdf
```

```python
dfsort = df.sort_values(by=['publish year'])
minyear = dfsort.iloc[0]['publish year']
maxyear = dfsort.iloc[-1]['publish year']


years=[i for i in range(minyear,maxyear+1)]
books = [0]*len(years)
bookread = [0]*len(years)
publishers = [set() for i in range(len(years))]
for i in dfsort.iloc:
    books[ i['publish year']-minyear]+=1
    bookread[i['publish year']-minyear]+= i['numRatings']
    publishers[i['publish year']-minyear].add(i['publisher'])


xpoints = years
ypoints = books
plt.ylabel("books")
plt.plot(xpoints, ypoints,color='m')
plt.show()



xpoints = years
ypoints = bookread
plt.ylabel("reading book")
plt.plot(xpoints, ypoints,color='g')
plt.show()


xpoints = years

ypoints = [ len(i) for i in publishers]
plt.ylabel("number of publishers")
plt.plot(xpoints, ypoints,color='r')
plt.show()
```

```python
sgdf = pd.DataFrame()                #seprate genres dataframe
for row in df.iloc:
    for genre in row['genres']:
        n =pd.DataFrame([row])
        n.pop('genres')
        n["genre"] = genre
        sgdf = pd.concat([sgdf,n])

sgdf = sgdf.reset_index().drop(['index'],axis=1)
sgdf
```

```python
newdf = pd.DataFrame( columns=['genre', 'numBooks', 'sumPages','numAuthors','firstBookYear','avgRating','sumRatings'])
newdf['genre'] = sgdf['genre'].unique()
newdf['numBooks'] = sgdf['genre'].value_counts(sort=False).values


newdf['sumPages'] = [sgdf.loc[sgdf['genre'] == i]['pages'].sum() for i in newdf.genre]
newdf['numAuthors'] = [sgdf.loc[sgdf['genre'] == i].author.nunique() for i in newdf.genre]
newdf['firstBookYear'] = [sgdf.loc[sgdf['genre'] == i]['publish year'].min() for i in newdf.genre]

newdf['avgRating'] = [sgdf.loc[sgdf['genre'] == i]['rating'].sum() for i in newdf.genre]
newdf['avgRating'] /= newdf['numBooks']

newdf['sumRatings'] = [sgdf.loc[sgdf['genre'] == i]['numRatings'].sum() for i in newdf.genre]

newdf
```

```python
n = 11 #show n first datas
fig, axs = plt.subplots(3, 2, figsize=(20, 20))
newdf.sort_values(by=['numBooks'],ascending=False,inplace=True)
x = newdf['genre'][0:n]
y = newdf['numBooks'][0:n]
axs[0, 0].bar(x, y,color='gray')
axs[0, 0].set_title('number of books')

newdf.sort_values(by=['sumPages'],ascending=False,inplace=True)
x = newdf['genre'][0:n]
y = newdf['sumPages'][0:n]
axs[0, 1].bar(x, y,color='red')
axs[0, 1].set_title('sum of pages')

newdf.sort_values(by=['numAuthors'],ascending=False,inplace=True)
x = newdf['genre'][0:n]
y = newdf['numAuthors'][0:n]
axs[1, 0].bar(x, y,color='darkviolet')
axs[1, 0].set_title('number of authors')



newdf.sort_values(by=['firstBookYear'],ascending=False,inplace=True)
x = newdf['genre'][0:n]
y = newdf['firstBookYear'][0:n]
axs[1, 1].set_ylim(min(y)-3, max(y)+3)
axs[1, 1].bar(x, y,color='aqua')
axs[1, 1].set_title('first Book Year')



newdf.sort_values(by=['avgRating'],ascending=False,inplace=True)
x = newdf['genre'][0:n]
y = newdf['avgRating'][0:n]
axs[2, 0].set_ylim(min(y)-0.2, max(y)+0.2)
axs[2, 0].bar(x, y,color='orange')
axs[2, 0].set_title('average Rating')


newdf.sort_values(by=['sumRatings'],ascending=False,inplace=True)
x = newdf['genre'][0:n]
y = newdf['sumRatings'][0:n]
axs[2, 1].bar(x, y,color='lime')
axs[2, 1].set_title('sum of ratings')

plt.show()
```
